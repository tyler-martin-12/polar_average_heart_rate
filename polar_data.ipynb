```python
df = pd.read_csv(file_name)
```

```python
times = df['Sport']
times = times[2:]
#times = times.reset_index()
#times = times.drop('index', axis = 1)
#times.columns = ['times']
```

```python
x = pd.DataFrame()
x['hr'] = hr
x['times'] = times
x.reset_index(inplace=True)
x.drop('index', axis=1, inplace=True)
x.head()
```

## 45 min stats

```python
num_seconds = 45 * 60
raw = x['hr'].apply(lambda y: float(y))
```

```python
for i in range(10):
    session = raw[i:i+num_seconds]
    print(session.mean())
```

```python
i + num_seconds
```

```python
can_continue = True
i = 0
max_mean = 0
while can_continue:
    session = raw[i:i + num_seconds]
    if session.mean() > max_mean:
        max_mean = session.mean()
    i += 1
    if (i + num_seconds) > len(raw):
        can_continue = False
print(max_mean)
```
